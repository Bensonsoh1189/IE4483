```python
import os
import csv
from ultralytics import YOLO
import numpy as np

# Load the custom model
model = YOLO('./runs/classify/train5/weights/last.pt')

# Specify the folder containing images
image_folder = './dataset/test'
output_csv = './predictions.csv'

# Prepare CSV to save results
with open(output_csv, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Filename', 'Predicted_Class', 'Probability'])

    # Loop through each image in the folder
    for image_name in os.listdir(image_folder):
        image_path = os.path.join(image_folder, image_name)
        
        # Ensure it is a file
        if os.path.isfile(image_path):
            # Predict on the image
            results = model(image_path)

            # Get class names and probabilities
            names_dict = results[0].names
            probs = results[0].probs.data.tolist()

            # Get the predicted class and probability
            predicted_class = names_dict[np.argmax(probs)]
            probability = max(probs)

            # Remove the file extension
            filename_without_extension = os.path.splitext(image_name)[0]

            # Write results to CSV
            writer.writerow([filename_without_extension, predicted_class, probability])

print(f"Predictions saved to {output_csv}")

```
